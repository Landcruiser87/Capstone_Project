```python
models = ['GRU', 'LSTM', 'BidirectionalGRU', 'BidirectionalLSTM', 'Conv1D', 'ConvLSTM2D', 'Dense']

sns.set_palette(sns.cubehelix_palette(8, start=.5, rot=-.75)) 

for model in models:
    subset = df[df['key'] == model]
    sns.distplot(subset['val_accuracy'], hist = False, label = model, norm_hist=False)
    
    
# Plot formatting

plt.legend(prop={'size': 9}, title = 'Model Type')
plt.title('Density Plot of Model Accuracies')
plt.xlabel('Validation Accuracy')
plt.ylabel('Density')
ax = plt.axes()
ax.yaxis.set_major_locator(plt.NullLocator())
```
